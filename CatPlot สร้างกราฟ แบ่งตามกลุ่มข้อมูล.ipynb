#### นำเข้า pandas

```python
import pandas as pd
```

#### เปิด csv

```python
tips = pd.read_csv('tips.csv')
tips.head(3)
```

#### นำเข้าเครื่องมื seaborn

```python
import seaborn as sns
```

#### ใช้ sns.catplot() สร้างกราฟเปรียบเทียบข้อมูลแต่ละกลุ่ม กำหนด x= ตั้ง column ที่ต้องใช้เป็นแกน x, y= ตั้ง column ที่ต้องการใช้เป็นแกน y, col= รับ column ที่ต้องการแบ่งข้อมูลเป็นคนละกราฟเพื่อเทียบ, hue= รับ column ที่ต้องการ แบ่งข้อมูลในกราฟ

```python
data = sns.catplot(x='day',y='tips_pct',data=tips,col='smoker',hue='sex',kind='box')
```
